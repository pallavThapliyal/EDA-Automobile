<h5>Replacing "?" with the mean<h5>

```python
ct = (data['normalized-losses'] =="?").sum()
ct
```

```python
d2= data['normalized-losses'].loc[data['normalized-losses']!="?"]
d2= d2.astype(int).mean()
d2
```

```python
data['normalized-losses'] = data['normalized-losses'].apply(lambda x : x.replace('?',"122"))
data['normalized-losses']
```

```python
for i in d:
    x= data[i] =="?"
    print(i,x.sum())
    
```

```python
d3= data['bore'].loc[data['bore']!="?"]
d3= d3.astype(str).astype(float).mean()
round(d3,2)
```

```python
data['bore'] = data['bore'].apply(lambda x : x.replace('?','3.33'))
data['bore']
```

```python
d4 = data['stroke'].loc[data['stroke']!="?"]
d4 = d4.astype(float).mean()
round(d4,2)
```

```python
data['stroke'] = data['stroke'].apply(lambda x : x.replace('?','3.26'))
data['stroke']
```

```python
d5 = data['horsepower'].loc[data['horsepower']!="?"]
d5 = d5.astype(int).mean()
round(d5,2)
```

```python
data['horsepower'] = data['horsepower'].apply(lambda x : x.replace('?','104.26'))
data['horsepower']
```

```python
d6 = data['peak-rpm'].loc[data['peak-rpm']!="?"]
d6 = d6.astype(int).mean()
round(d6,2)
```

```python
data['peak-rpm'] = data['peak-rpm'].apply(lambda x : x.replace('?','5125.37'))
data['peak-rpm']
```

```python
d7 = data['price'].loc[data['price']!="?"]
d7 = d7.astype(int).mean()
round(d7,2)
```

```python
data['price'] = data['price'].apply(lambda x : x.replace('?','13207.13'))
data['price']=data["price"].astype(float)
data['price']
```

```python
data['num-of-doors'] = data['num-of-doors'].loc[data['num-of-doors']!='?']
data.isnull().sum()
data.dropna(inplace= True)
data.head(30)
```
